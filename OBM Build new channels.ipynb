```python
import os
import cv2
import numpy as np
from matplotlib import pyplot
from channel import Channel
# Colors (B, G, R)
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)

def create_blank(width, height, color=(0, 0, 0)):
    """Create new image(numpy array) filled with certain color in BGR"""
    image = np.zeros((height, width, 3), np.uint8)
    # Fill image with color
    image[:] = color
    return image
```

```python
# Create new blank 300x300 black image
width, height = 300, 300
image = create_blank(width, height, color=BLACK)

# define params of Channel object
params = {
    "angdev": 15, # stdev 15 degrees in center line angle delta random var
    "seg_hi": 30, # upper value of segment length random var
    "seg_lo": 25, # lower value of segment length random var
    "amp_hi": 10, # upper value of amplitute random var
    "amp_lo": 2, # lower value of amplitude random var
    "num_channels": 3, # Number of channels to draw
}

# draw set of channels from South to North
num_channels = params["num_channels"]
# padding between channel sources
padding = width // (num_channels + 1)

for i in range(padding, width, padding):
    # create channel
    chObj = Channel(image, (i, 0), 0, angle_dev=params["angdev"], 
                    segment_hi=params["seg_hi"], segment_lo=params["seg_lo"], 
                    amp_hi=params["amp_hi"], amp_lo=params["amp_lo"], 
                    thickness=2)
    # draw channel segments
    for i in range(height // (params["seg_lo"] // 2)):
        chObj.add_segment()

# convert to monochrome
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
pyplot.figure(figsize=image.shape)
im = pyplot.imshow(image)
pyplot.gca().invert_yaxis()
pyplot.colorbar(im)
pyplot.show()
```
